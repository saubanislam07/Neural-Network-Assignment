```python
# 1. Load the "Train" folder (The Flashcards)
train_folder = '/content/drive/MyDrive/Neural_Network/X-Ray/train'
training_pile = tf.keras.utils.image_dataset_from_directory(
    train_folder,
    image_size=(150, 150),
    batch_size=32
)

# 2. Load the "Val" folder (The Practice Quiz)
val_folder = '/content/drive/MyDrive/Neural_Network/X-Ray/val'
practice_pile = tf.keras.utils.image_dataset_from_directory(
    val_folder,
    image_size=(150, 150),
    batch_size=32
)

# 3. Load the "Test" folder (The Final Exam)
test_folder = '/content/drive/MyDrive/Neural_Network/X-Ray/test'
exam_pile = tf.keras.utils.image_dataset_from_directory(
    test_folder,
    image_size=(150, 150),
    batch_size=32
)

print("All pictures are loaded correctly and ready!")
```

```python
# 1. We start with a blank brain (a Sequential model means we stack blocks one by one)
robot_brain = tf.keras.models.Sequential([

    # Block 1: The first magnifying glass (looks for simple edges and lines)
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2), # This shrinks the picture to make it easier to process

    # Block 2: A stronger magnifying glass (looks for bigger shapes, like spots or shadows)
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    # Block 3: Laying out the puzzle pieces
    # The brain needs the clues in a straight line before it makes a final decision
    tf.keras.layers.Flatten(),

    # Block 4: The Decision Maker!
    # A 'Dense' layer with 1 node and a 'sigmoid' activation is perfect for a
    # simple "Yes or No" (Sick or Healthy) guess.
    tf.keras.layers.Dense(1, activation='sigmoid')
])

print("The Robot Brain is built and ready to learn!")
```

```python
# We start with a blank brain
robot_brain = tf.keras.models.Sequential([

    # NEW: The Welcome Mat! It tells the brain the exact size of the pictures coming in
    tf.keras.layers.Input(shape=(150, 150, 3)),

    # Block 1: The first magnifying glass
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    # Block 2: A stronger magnifying glass
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    # Block 3: Laying out the puzzle pieces into a straight line
    tf.keras.layers.Flatten(),

    # Block 4: The Decision Maker! (Sick or Healthy)
    tf.keras.layers.Dense(1, activation='sigmoid')
])

print("The Robot Brain is built with the new Welcome Mat!")
```

```python
# Giving the robot the rules of the game
robot_brain.compile(
    optimizer='adam',             # 'adam' is a very smart strategy the robot uses to learn quickly
    loss='binary_crossentropy',   # This is the math rule for a "Yes or No" (Sick or Healthy) guessing game
    metrics=['accuracy']          # We want the computer to show us its score as a percentage (like 85% correct)
)

print("The rules are set! Let's play.")
```

```python
# Start the timer and begin studying!
print("Starting the flashcards...")

history = robot_brain.fit(
    training_pile,                  # 1. The main flashcards to learn from (Train folder)
    epochs=5,                       # 2. Go through the whole pile 5 times
    validation_data=practice_pile   # 3. Take a practice quiz after every round (Val folder)
)

print("Phew! The robot is done studying.")
```

```python
# 1. Give the robot the Final Exam
print("Handing out the final exam...")
test_loss, test_accuracy = robot_brain.evaluate(exam_pile)

# 2. Print out the final grade
final_grade = test_accuracy * 100
print(f"The robot's Final Exam Score is: {final_grade}%")
```

```python
# 3. Save the robot into a file
robot_brain.save('my_pneumonia_robot.keras')

print("Robot safely packed up into a file! You can now download it.")
```

```python
# Building a smarter brain to stop memorizing!
robot_brain_v2 = tf.keras.models.Sequential([

    # The Welcome Mat
    tf.keras.layers.Input(shape=(150, 150, 3)),

    # Block 1: First magnifying glass
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    # Block 2: Second magnifying glass
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    # NEW Block 3: Third, deepest magnifying glass!
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),

    # Laying out the puzzle pieces
    tf.keras.layers.Flatten(),

    # NEW: The Blindfold!
    # '0.5' means we turn off 50% of the brain while it studies
    tf.keras.layers.Dropout(0.5),

    # The Decision Maker
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Tell this new brain the rules of the game
robot_brain_v2.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print("Robot Brain Version 2.0 is built and ready for a rematch!")
```

```python
# 1. Start studying with the new brain (10 rounds this time!)
print("Starting the flashcards for Version 2.0...")
history_v2 = robot_brain_v2.fit(
    training_pile,
    epochs=10,
    validation_data=practice_pile
)

# 2. Hand out the Final Exam
print("Handing out the final exam...")
test_loss_v2, test_accuracy_v2 = robot_brain_v2.evaluate(exam_pile)

# 3. Print out the new final grade
final_grade_v2 = test_accuracy_v2 * 100
print(f"Version 2.0 Final Exam Score is: {final_grade_v2}%")

# 4. Save the smarter robot into a new file
robot_brain_v2.save('my_smarter_pneumonia_robot.keras')
print("Smarter robot safely packed up into a file!")
```
